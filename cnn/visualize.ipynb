```python
%config InlineBackend.figure_format='retina'

import typing as ty

from matplotlib import pyplot as plt
from matplotlib import ticker as mticker
import pandas as pd
```

```python
RESULTS: dict[str, tuple[str, str]] = {
    "baseline": ("Baseline", "data/result.baseline.ablated.json"),
    "ours": ("Ours", "data/result.json"),
    "ours_nose": ("Ours (no SE)", "data/result.nose.ablated.json"),
    "ours_nores": ("Ours (no Res)", "data/result.nores.ablated.json"),
    "ours_noseres": ("Ours (no SE, no Res)", "data/result.noseres.ablated.json"),
}

RESULTS_PD = {r: pd.read_json(RESULTS[r][1], typ="frame").T for r in RESULTS}
```

## Results

```python
for r in ("ours", "baseline"):
    RESULTS_PD[r]["test_loss"].plot(label=RESULTS[r][0])
plt.gca().xaxis.set_major_locator(mticker.MaxNLocator(integer=True))
plt.xlabel("Epoch")
plt.ylabel("Test loss")
plt.legend()
plt.show()
```

```python
for r in ("ours", "baseline"):
    RESULTS_PD[r]["acc"].plot(label=RESULTS[r][0])
plt.gca().xaxis.set_major_locator(mticker.MaxNLocator(integer=True))
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

### Ablation study

```python
for r in ("ours", "ours_nose", "ours_nores", "ours_noseres"):
    RESULTS_PD[r]["test_loss"][:17].plot(label=RESULTS[r][0])
plt.gca().xaxis.set_major_locator(mticker.MaxNLocator(integer=True))
plt.xlabel("Epoch")
plt.ylabel("Test loss")
plt.legend()
plt.show()
```
